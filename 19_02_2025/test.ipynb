```python
import pandas as pd
import numpy as np
```

```python
product_df=pd.read_csv(r"C:\Users\20368750\OneDrive - Larsen & Toubro\Documents\Digital Initiative\DSA_PRACTICE\19_02_2025\products.csv")
product_df.head()
```

```python
sales_df=pd.read_csv(r"C:\Users\20368750\OneDrive - Larsen & Toubro\Documents\Digital Initiative\DSA_PRACTICE\19_02_2025\sales_data.csv")
sales_df.head()
```

```python
combined_data = pd.merge(product_df, sales_df, on='product_id', how='inner')
combined_data
```

```python
combined_data.shape
```

```python
combined_data['total_sales']=combined_data['quantity']*combined_data['price']
combined_data
```

```python
combined_data['date']=pd.to_datetime(combined_data['date'])
combined_data
```

```python
filtered_data=combined_data[combined_data['date'].dt.year==2023]
filtered_data
```

```python
category_sales =filtered_data.groupby('category').agg(total_quantity=('quantity', 'sum'),
    total_sales=('price', 'sum') 
).reset_index()

category_sales
```

```python
top_category = category_sales.loc[category_sales['total_sales'].idxmax()]
top_category
# top_category=category_sales.iloc[category_sales['total_sales'].idmax()]
# top_category
```

```python
combined_data['log_price']=np.log(combined_data['price'])
combined_data
```

```python
mean_sales=combined_data['total_sales'].mean()
dev_sales=combined_data['total_sales'].std()
print(mean_sales)
print(dev_sales)
```

```python
combined_data['sales_zscore']=(combined_data['total_sales']-mean_sales)/(dev_sales)
combined_data
```

```python
missing_qty = combined_data[combined_data['quantity'].isnull()]
missing_qty
```
